```python
import re
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
```

```python
sent_tokenized_data = sent_tokenize(corpus)
sent_tokenized_data
```

```python
preprocessed_data = []
for sent in sent_tokenized_data:
     text = re.sub('[^a-zA-Z0-9]',' ', sent)
     text = text.lower()
     text = word_tokenize(text)

     lemmatizer = WordNetLemmatizer()
     lemmatized_text = [lemmatizer.lemmatize(t) for t in text if t not in stopwords.words('english') ]
     # lemmatized_text = ' '.join(lemmatized_text)
     preprocessed_data.append(lemmatized_text)
```

```python
def question_preprocessing(question):
     q_t = re.sub('[^a-zA-Z0-9]',' ', question)
     q_t = q_t.lower()
     q_t = word_tokenize(q_t)

     lemmatizer = WordNetLemmatizer()
     lemmatized_text = [lemmatizer.lemmatize(t) for t in q_t if t not in stopwords.words('english') ]
     lemmatized_text = ' '.join(lemmatized_text)
     return lemmatized_text
```

```python
model = Word2Vec.load("word2vec_model.bin")
model
```

```python
def get_similar_words(question, model, top_n=5):
    # Tokenize the question into words
    words = question.split()
   
    # Get similar words for each word in the question
    similar_words = {}
    for word in words:
        if word in model.wv:
            similar = model.wv.most_similar(word, topn=top_n)
            similar_words[word] = similar
   
    return similar_words
```
